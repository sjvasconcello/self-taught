## Example 2 - Multiple Minimal vs Initial Guess & Adanced Function
$$
g(x) = x^4 - 4x^2 + 5
$$

```python
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

## Gradient Descent as a Python Function

```python
def gradient_descent(derivative_func,initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    new_x = initial_guess  
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x =  previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, 
                                                 initial_guess=-0.5,
                                                 multiplier=0.01,
                                                 precision=0.001)
print(f"Local min occurs at: {local_min}")
print(f"Number od steps: {len(list_x)}")
```

```python
# ling gradient descenttfunction


local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, 
                                                 initial_guess=0)

plt.figure(figsize=[15,5])

plt.subplot(1,2,1)

plt.xlim(-2,2)
plt.ylim(0.5,5.5)

plt.title("Cost function", fontsize=17)
plt.xlabel("X",fontsize=16)
plt.ylabel("g(x)", fontsize=16)

plt.plot(x_2,g(x_2), color="blue", linewidth=3, alpha=0.8)
plt.scatter(list_x, g(np.array(list_x)), color="red", s=100, alpha=0.6)

# Derivative
plt.subplot(1,2,2)

plt.title("Slope of the cost functin", fontsize=17)
plt.xlabel("X",fontsize=16)
plt.ylabel("dg(x)", fontsize=16)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6,8)

plt.plot(x_2,dg(x_2), color="skyblue", linewidth=3, alpha=0.6)
plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.5)


plt.show()
```

## Batch Gradient Decent with Sympy

```python
# Setup
multilpier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = diff(f(a,b),a).evalf(subs={a: params[0], b: params[1]})
    gradient_y = diff(f(a,b),b).evalf(subs={a: params[0], b: params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multilpier * gradients

print(f"Values in gradient array {gradients}")
print(f"Minimum values occurs at x value of: {params[0]}")
print(f"Minimum values occurs at y value of: {params[1]}")
print("The cost is: ", f(params[0], params[1]) )
```

```python
# Partial derivative functions example 4

def fpx(x, y):
    r = 3**(-x**2 - y**2)
    return 2*x*log(3)*r / (r+1)**2

def fpy(x, y):
    r = 3**(-x**2 - y**2)
    return 2*y*log(3)*r / (r+1)**2
```

```python
fpx(1.8, 1.0)
```

```python
# Setup
multilpier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = fpx(params[0],params[1])
    gradient_y = fpy(params[0],params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multilpier * gradients

print(f"Values in gradient array {gradients}")
print(f"Minimum values occurs at x value of: {params[0]}")
print(f"Minimum values occurs at y value of: {params[1]}")
print("The cost is: ", f(params[0], params[1]) )
```

## Example 5 - Working with Data & a Real Cost Function

### Mean Square Error: a cost function for regression problems

### $$ RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$

### $$ MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$

### $$ MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - \hat{y} \big)^2 $$

```python
# Make sample data
x_5 = np.array([[0.1,1.2,2.4,3.2,4.1,5.7,6.5]]).T
y_5 = np.array([1.7,2.4,3.5,3.0,6.1,9.4,8.2]).reshape(7,1)

print("Shape of x_5 array: ", x_5.shape)
print("Shape of y_5 array: ", y_5.shape)
```

```python
# Quick linear regression
regr = LinearRegression()
regr.fit(x_5,y_5)
print("Theta 0: ", regr.intercept_[0])
print("Theta 1: ", regr.coef_[0][0])
```

```python
plt.scatter(x_5, y_5, s=50)
plt.plot(x_5, regr.predict(x_5),color="orange", linewidth=3)
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# y_hat = theta0 + theta1*x
y_hat = 0.8475351486029545 + 1.2227264637835913*x_5
print(f"Est values y_hat are: \n {y_hat} \n")
print("In comparison, the actual y values are \n", y_5)
```

```python
def mae(y, y_hat):
    mse_calc = 1/y.size * sum((y - y_hat)**2)
    return mse_calc

mse(y_5, y_hat)
```

```python
mean_squared_error(y_5,regr.predict(x_5))
```

## Partial Deriatives of MSE w.r.t $\theta_0$ and $\theta_1$

### $$ \frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum^{n}_{i=1} (y^{(i)} - \theta_0 - \theta_1 x^{(i)}) $$

### $$ \frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum^{n}_{i=1} (y^{(i)} - \theta_0 - \theta_1)(x^{(i)}) $$

### MSE & Gradient Descent

```python
def grad(x,y,theta):
    n = y.size
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - theta[0] - theta[1]*x)*x)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9,2.9])

plot_vals = thetas.reshape(1,2)
mse_vals =  mae(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5,y_5,thetas)
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1,2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mae(y_5,thetas[0] + thetas[1]*x_5))
    
print("Min occurs at Theta 0: ", thetas[0])
print("Min occurs at Theta 1: ", thetas[1])
print("MSE is: ", mae(y_5, thetas[0] + thetas[1]*x_5))
```
